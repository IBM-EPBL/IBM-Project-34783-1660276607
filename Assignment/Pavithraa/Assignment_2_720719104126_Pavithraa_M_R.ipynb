```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Geography'] = le.fit_transform(df['Geography'])
```

```python
df['Surname'] = le.fit_transform(df['Surname'])
```

```python
df['Gender']= le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
x = df.iloc[:,0:13].values
y = df.iloc[:,13:14].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train.shape , x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
regressor = Sequential()
regressor.add(Dense(13,activation='relu'))
regressor.add(Dense(20,activation='relu'))
regressor.add(Dense(18,activation='relu'))
regressor.add(Dense(19,activation='relu'))
regressor.add(Dense(1,activation='linear'))
```

```python
regressor.compile(optimizer='adam',loss='mse',metrics=('mse'))
```

```python
regressor.fit(x_train,y_train,batch_size=50,epochs=300)
```

```python
ypred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,ypred)*100
```
